## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

A scatter plot identifies a possible relationship between changes observed in two or more different sets of variables. It provides a visual and statistical means to test the strength of a relationship between two or three variables.

```python
import numpy as np
```

```python
np.random.seed(3)
```

```python
x = 3 + np.random.normal(0, 2, 50)
```

```python
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph between X and Y")
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

subplot() function adds subplot to a current figure at the specified grid position.  It is similar to the subplots() function however unlike subplots() it adds one subplot at a time. So to create multiple plots you will need several lines of code with the subplot() function.

```python
import numpy as np
```

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
```

```python
fig = plt.figure()
ax = fig.add_subplot()
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph for line 1 ")
ax.plot(x,y, c= 'b')
```

```python
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([50, 20, 40, 20, 60, 70])
```

```python
fig = plt.figure()
ax1 = fig.add_subplot()
ax1.plot(x1,y1, c = 'r', alpha = 0.9)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph for line 2 ")
plt.show()
```

```python
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x3,y3, c = '#0DF210')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph for line 3")
plt.show()
```

```python
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x4,y4, c = '#EFF704')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph for line 4")
plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories. One of the axis of the plot represents the specific categories being compared, while the other axis represents the measured values corresponding to those categories.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Vertical Bar Graph")
```

```python
plt.barh(company, profit)
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Horizontal Bar Graph")
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median. Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
box = [box1, box2]
```
